# bad: drop entire rows if a single column has a NaN in it

https://www.dhs.gov/tsa-claims-data

```python
df = pandas.read_excel('https://www.dhs.gov/sites/default/files/publications/claims-2014.xls')
```

```python
df.shape
```

```python
numpy.count_nonzero(df.isnull())
```

```python
df.head(8)
```

```python
df =df.replace(r'-', numpy.nan, regex=True)
```

how many cells have NaN?

```python
numpy.count_nonzero(df.isnull())
```

out of a total of

```python
df.shape[0]*df.shape[1]
```

so slightly more than 10% of the cells have NaN

What happens if we drop any row that has a NaN?

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html

print('before:',df.shape)
df=df.dropna(how='any',axis='rows')
print('after: ',df.shape)
```

Dropping more than half the rows may be sufficient to introduce error in the claims, especially if the dropped rows have shared features

# bar chart ordering

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.bar.html

```python
df = pandas.DataFrame({'labl':['A', 'B', 'C','D','E','F'], 'valu':[10, 30, 20,50,2,5]})
df
```

```python
ax = df.plot.bar(x='labl', y='valu', rot=0)
ax.get_legend().remove()
plt.show()
```

```python
df.sort_values('valu', ascending=False)
```

```python
_=df.sort_values('valu', ascending=False).plot.bar(x='labl', y='valu',rot=0)
```
