https://stackoverflow.com/questions/44867219/pandas-filling-na-values-to-be-filled-based-on-distribution-of-existing-values

```python
series_to_draw_from = df['value'].dropna()
```

```python
for row_indx in range(df.shape[0]):
    if(numpy.isnan(df.at[row_indx,'value'])):
        df.at[row_indx,'value'] = random.choice(list(series_to_draw_from))
```

```python
plt.plot_date(df['date'],df['value']);
```
