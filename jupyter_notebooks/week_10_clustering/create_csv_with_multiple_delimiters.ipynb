```python
import pandas
import numpy
import requests
import random
```

```python
# https://stackoverflow.com/questions/18834636/random-word-generator-python

word_site = "http://svnweb.freebsd.org/csrg/share/dict/words?view=co&content-type=text/plain"

response = requests.get(word_site)
WORDS = response.content.splitlines()
```

```python
word_list = [x.decode('utf-8') for x in WORDS]
word_list[0:10]
```

```python
row_ct = 1000
col_ct = 40
```

```python
col_list=[]
for indx in range(col_ct):
    col_list.append(random.choice(word_list))
```

```python
df = pandas.DataFrame(numpy.random.randint(0,100,size=(row_ct, col_ct)),columns=col_list)
df.shape
```

```python
df.head()
```

```python
for col in df.columns:
    for indx in range(df.shape[0]//2):
        df.loc[df.sample(1).index,col] = random.choice(word_list)
```

```python
df.head()
```

```python
with open('file_to_parse.csv','w') as fil:
    for col in df.columns[:-1]:
        fil.write(col+', ')
    fil.write(df.columns[-1])
    fil.write("\n")
    for indx in range(df.shape[0]):
        fil.write(str(df.at[indx,df.columns[0]])+" : ")
        for col in range(1,df.shape[1]-1):
            fil.write(str(df.at[indx,df.columns[col]])+" | ")
        fil.write(str(df.at[indx,df.columns[df.shape[1]-1]]))
        fil.write("\n")
```
